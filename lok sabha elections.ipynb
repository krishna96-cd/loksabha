```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def get_best_matches(list1, list2, threshold=80):
    matches = {}
    for item in list1:
        best_match = process.extractOne(item, list2, scorer=fuzz.token_sort_ratio)
        if best_match[1] >= threshold:
            matches[item] = best_match[0]
        else:
            matches[item] = None
    return matches

# Get best matches
matches = get_best_matches(gdp_14_li, df_li_state)  

# Print the matches
trans_state_li = []
for key, value in matches.items():
    print(f'{key} -> {value}')
    trans_state_li.append(key)
    
    
    

```
